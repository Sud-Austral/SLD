```python
#Autor: Ian Meza
#Fecha: 2/Julio/2021
#Descripción: Secuencia para crear un archivo de estilos para GeoServer en formato .sld 

#<----- Librería ----->
import pandas as pd

#<----- Lectura de archivos necesarios ----->
df = pd.read_excel("Datos_Censo_2017_Comuna.xlsx")
paletas = pd.read_excel("paletas.xlsx")

a=[]
with open('Plantilla_sld.txt', 'r') as f:
    a = f.read()
f.closed

#<----- Columna a aplicar rangos ----->
#df["TOTAL_PERS"].max()
#df["TOTAL_PERS"].min()
diferencia = df["TOTAL_PERS"].max() - df["TOTAL_PERS"].min()
div = diferencia/5

#<----- Separación de bloques de datos ----->
inicio = a.split("<se:Rule>")[0]
#.replace("Datos_Censo_2017_Comuna copiar", "Nombre_de_tabla")
final = a.split("<se:Rule>")[1].split("</se:Rule>")[1]
referencia = "\n        <se:Rule>"+a.split("<se:Rule>")[1]

#<----- Creación de rangos ----->
r1min = df["TOTAL_PERS"].min()
print("Rango 1 mínimo")
print(r1min)
r1max = div+df["TOTAL_PERS"].min()+0.001
print("Rango 1 máximo")
print(r1max)

r2min = r1max+0.001
print("Rango 3 mínimo")
print(r2min)
r2max = r1max+div
print("Rango 3 máximo")
print(r2max)

r3min = r2max+0.001
print("Rango 3 mínimo")
print(r3min)
r3max = r2max+div
print("Rango 3 máximo")
print(r3max)

r4min = r3max+0.001
print("Rango 4 mínimo")
print(r4min)
r4max = r3max+div
print("Rango 4 máximo")
print(r4max)

r5min = r4max+0.001
print("Rango 5 mínimo")
print(r5min)
r5max = df["TOTAL_PERS"].max()
print("Rango 5 máximo")
print(r5max)

r1 = str(r1min)+" - "+str(r1max)
r2 = str(r2min)+" - "+str(r2max)
r3 = str(r3min)+" - "+str(r3max)
r4 = str(r4min)+" - "+str(r4max)
r5 = str(r5min)+" - "+str(r5max)

rangos = [r1,r2,r3,r4,r5]
minimos = [r1min,r2min,r3min,r4min,r5min]
maximos = [r1max,r2max,r3max,r4max,r5max]

#Ciclo de cración del XML

salida = []
l = 0
for i in rangos:
    ref = referencia.replace("<se:Name>6444 - 65152</se:Name>", "<se:Name>"+i+"</se:Name>")
    ref = ref.replace("<se:Title>6444 - 65152</se:Title>", "<se:Title>"+i+"</se:Title>")
    j = 0
    k = 0
    ref = ref.replace("<ogc:Literal>6444</ogc:Literal>", "<ogc:Literal>"+str(j)+"</ogc:Literal>")
    ref = ref.replace("<ogc:Literal>65151.80000000001018634</ogc:Literal>", "<ogc:Literal>"+str(k)+"</ogc:Literal>")
    ref = ref.replace("fff5eb", paleta[l])
    j = j+1
    k = k+1
    l = l+1
    salida.append(ref)
    
print(salida)
```
