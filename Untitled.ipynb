```python
region = []
variables = []
rmax = []
rmin = []
limI = []
limS = []
for i in df["COMUNA"].unique():
    df_aux = df[df["COMUNA"] == i ]
    for j in df.columns[8:-2]:
        
        print(f"Para la región {i} para la variable {j} tenemos los siguientes límites")
        region.append(i)
        variables.append(j)
        print(f"Limite inferior {df_aux[j].min()}")
        print(f"Limite superior {df_aux[j].max()}")
        rmin.append(df_aux[j].min())
        rmax.append(df_aux[j].max())
        for k in rmin:
            for l in rmax:
                diferencia = l - k
                div = diferencia/5
                ri = k
                rs = k+div
                limI.append(ri)
                limS.append(rs)
```

```python
tabla = pd.DataFrame({"ID_REGION": region, "VARIABLES": variables, "LÍMITE MÍNIMO": rmin, "LÍMITE MÁXIMO": rmax})
tabla.to_excel("Límites_Rangos_Comuna.xlsx", index= False)
```
