```python
ruta = r"C:\Users\Ian\Documents\Github\SLD"
contenido = os.listdir(ruta)
archivos = []
for fichero in contenido:
    if os.path.isfile(os.path.join(ruta, fichero)) and fichero.endswith('.sld'):
        archivos.append(fichero)
print(archivos)
```

```python
#Region

numero = 0
letra = "a"

for i in archivos:
    
    nombre = i[:-4]
    salida = a.replace("region_1_TOTAL_PERS", f"{nombre}")
    for i in range(200):
        salida = salida.replace("i86", letra+"{:0>2}".format(numero))
        #print(letra+"{:0>2}".format(numero))
        numero = numero+1
        if(numero==100):
            numero = 0
            letra = "b"
    text_file = open(nombre+".xml", "w")

    text_file.write(salida)

    text_file.close()
```
